# LINE CHART

```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
df = pd.read_csv('/kaggle/input/sandp500/all_stocks_5yr.csv')
df.head(10)
```

```python
# Load the magical CSV file that holds the secrets of the stock market
file_path = '/kaggle/input/sandp500/all_stocks_5yr.csv'
df = pd.read_csv(file_path)

# Give the date column a little makeover, turning it into a fancy datetime format
df['date'] = pd.to_datetime(df['date'])

# Focus on a few stocks, like our favorite superheroes—let's say 'AAPL', 'GOOGL', and 'MSFT'
selected_stocks = ['AAPL', 'GOOGL', 'MSFT']  
filtered_df = df[df['Name'].isin(selected_stocks)]

# Wave our Plotly wand to create a mesmerizing line chart
# Different colors for each stock, making the closing prices dance over time
fig = px.line(filtered_df, x='date', y='close', color='Name', title='Stock Prices Over Time', labels={'close': 'Closing Price', 'date': 'Date'})

# Lets Show off our creation
fig.show()
```

 
# BAR CHART

```python
file_path2 = '/kaggle/input/amazon-sales-dataset/amazon.csv'
df2 = pd.read_csv(file_path2)
df2.head()
```

```python
# Load the treasure trove of Amazon sales data
file_path2 = '/kaggle/input/amazon-sales-dataset/amazon.csv'
df = pd.read_csv(file_path2)

# Give the rating column a little nudge to become numeric
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Group by category and calculate the average rating - let's see which categories are getting all the love!
category_avg_rating = df.groupby('category')['rating'].mean().reset_index()

# If we want to limit to the top 200 categories we can do this:
category_avg_rating = category_avg_rating.head(200)  

# Truncate the category names to a maximum length, say 15 characters so that it is easy to understand
category_avg_rating['short_category'] = category_avg_rating['category'].apply(lambda x: x[:15] + '...' if len(x) > 15 else x)

# It's like a popularity contest for Amazon products!
fig = px.bar(category_avg_rating, x='short_category', y='rating', title='Average Rating by Product Category', labels={'rating': 'Average Rating', 'short_category': 'Product Category'})

# Lets Show off our chart - time to see who the real MVPs of Amazon!
fig.show()
```


# Pie Chart 

```python
# Loading the e-commerce treasure chest
file_path3= '/kaggle/input/ecommerce-behavior-data-from-multi-category-store/2019-Nov.csv'
df2 = pd.read_csv(file_path3)
df2.head()
```

```python
# Counting the occurrences of each event type, let's see what shoppers are up to!
event_counts = df2['event_type'].value_counts().reset_index()
event_counts.columns = ['event_type', 'count']

# Who knew shopping could be this colorful?
fig = px.pie(event_counts, names='event_type', values='count', title='Distribution of Event Types on E-commerce Platform')

# Who doesn't love a good slice of data?
fig.show()
```


# Scatter Plot

```python
# Loading Healthy Peoples
file_path5= '/kaggle/input/heights-and-weights-dataset/SOCR-HeightWeight.csv'
df2 = pd.read_csv(file_path5)
df2.head()
```

```python
# Convert height from inches to centimeters and weight from pounds to kilograms
df2['Height(cm)'] = df2['Height(Inches)'] * 2.54
df2['Weight(kg)'] = df2['Weight(Pounds)'] * 0.453592

# Scatter plots are like connect-the-dots for grown-ups!
fig = px.scatter(df2, x='Height(cm)', y='Weight(kg)', title='Relationship Between Height and Weight')

# Update layout with custom axis titles - because even axes deserve some love!
fig.update_layout(xaxis_title='Height (cm)', yaxis_title='Weight (kg)')

# Show off our scatter plot - it's a weighty matter of height!
fig.show()
```


# Area Chart

```python
# Loading the ultimate gaming sales dataset
file_path6= '/kaggle/input/videogamesales/vgsales.csv'
df = pd.read_csv(file_path6)
# Sneak peek at the first few rows of the dataset - because we love spoilers!
df.head()
```

```python
# Let's group by Year and sum up the Global Sales to see the trend over the years
sales_per_year = df.groupby('Year')['Global_Sales'].sum().reset_index()

# Creating the area chart - it's like painting a beautiful landscape of sales!
fig = px.area(sales_per_year, x='Year', y='Global_Sales', title='Global Sales of Video Games Over the Years', 
              labels={'Global_Sales': 'Global Sales (in millions)', 'Year': 'Year'})

# Show off our area chart, its mountain of sales!
fig.show()
```


# Histograms

```python
import numpy as np

# Generating sample data: Exam scores
np.random.seed(0)
scores = np.random.normal(70, 10, 1000)  # Which Mean of 70, standard deviation of 10, 1000 samples

# Creating a histogram
fig = px.histogram(scores, nbins=30, title='Distribution of Exam Scores')
fig.update_layout(xaxis_title='Exam Scores', yaxis_title='Count')
fig.show()
```


# Box Plot

```python
# Loading the secrets of the world
file_path7= '/kaggle/input/employee-salaries-analysis/Employee_Salaries.csv'
df = pd.read_csv(file_path7)
# Sneaking into unauthorized territory
df.head()
```

```python
# Abbreviate the department names using slicing (first 10 characters)
df['Short_Department_Name'] = df['Department_Name'].str.slice(0, 16)

# Creating the box plot - let's compare the salary distributions across different departments!
fig = px.box(df, x='Short_Department_Name', y='Base_Salary', 
             title='Distribution of Salaries Across Different Departments',
             labels={'Base_Salary': 'Base Salary (in USD)', 'Short_Department_Name': 'Department'})

# Show off our box plot - because comparing salaries is like comparing apples to oranges, but in a good way!
fig.show()
```
